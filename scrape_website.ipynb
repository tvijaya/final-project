```python
def getBreachesData(url, year):
    df = pd.read_html(url)
    breaches_df = df[0]
    file_name = 'breaches_' + year + '.csv'
    breaches_df.to_csv(file_name)
    print(len(breaches_df))
```

```python
url_2017 = 'http://breachlevelindex.com/data-breach-database'
url_2016 = "http://breachlevelindex.com/data-breach-database.php?range=2016"
url_2015 = "http://breachlevelindex.com/data-breach-database.php?range=2015"
url_2014 = "http://breachlevelindex.com/data-breach-database.php?range=2014"
url_2013 = "http://breachlevelindex.com/data-breach-database.php?range=2013"

getBreachesData(url_2017, "2017")
getBreachesData(url_2016, "2016")
getBreachesData(url_2015, "2015")
getBreachesData(url_2014, "2014")
getBreachesData(url_2013, "2013")

```

```python
def getBreachesDataLength(url):
    df = pd.read_html(url)
    breaches_df = df[0]
    return len(breaches_df)
```
